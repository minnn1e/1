```python
import pandas as pd
cnts = pd.read_csv('read-counts.txt',sep='\t',comment='#',index_col=0)
cnts.head()
```

```python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
```

```python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1,1,figsize=(5,5))
ax.scatter(np.log2(cnts['clip_enrichment']),
           np.log2(cnts['rden_change']))
```

```python
from scipy.stats import pearsonr

fig, ax = plt.subplots(1, 1, figsize=(5, 5))

x = np.log2(cnts['clip_enrichment'])
y = np.log2(cnts['rden_change'])

r, p_value = pearsonr(x,y)

ax.scatter(x,y,color='black',s=10)

# 축 라벨 설정 (논문 Figure 4D 참고)
ax.set_xlabel('log2(clip_enrichment)')
ax.set_ylabel('log2(rden_change)')  # y축 라벨 변경

ax.text(0.95,0.05,f'r = {r:.2f}',transform=ax.transAxes, ha='right',va='bottom')


# 그래프 표시
plt.show()
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
mouselocal = pd.read_csv('https://hyeshik.qbio.io/binfo/mouselocalization-20210507.txt', sep='\t')
mouselocal.head()
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import pearsonr

# mouselocal 데이터프레임의 'Gene name' 열 이름을 'Gene names'로 변경
mouselocal = mouselocal.rename(columns={'Gene name': 'Gene names'})

# cnts 데이터프레임의 'Geneid' 열을 'Gene names'로 변경
cnts = cnts.rename(columns={'Geneid': 'Gene names'})

# 'Gene names' 열을 데이터 열로 되돌리기, 기존 'level_0' 열이 있다면 삭제
# drop=True를 추가하여 기존 index를 제거합니다.
cnts = cnts.reset_index(drop=True)

# 'Gene names' 열의 대소문자를 통일합니다.
cnts['Gene names'] = cnts['Gene names'].str.upper()
mouselocal['Gene names'] = mouselocal['Gene names'].str.upper()

# cnts와 mouselocal data 병합, clip_enrichment와 rden_change 열을 포함
# validate='one_to_many'를 추가하여 병합 작업을 검증합니다.
merged_df = pd.merge(cnts[['Gene names', 'clip_enrichment', 'rden_change']],
                    mouselocal, on='Gene names', how='inner', validate='one_to_many')

# localization category별 데이터 분류
cyto_df = merged_df[merged_df['Major location'] == 'Cytoplasm']
nuc_df = merged_df[merged_df['Major location'] == 'Nucleus']
ribo_df = merged_df[merged_df['Major location'] == 'Ribosome']

# scatter plot 생성
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

x = np.log2(merged_df['clip_enrichment'])
y = np.log2(merged_df['rden_change'])
r, p_value = pearsonr(x, y)

ax.scatter(np.log2(cyto_df['clip_enrichment']), np.log2(cyto_df['rden_change']), label='Cytoplasm', s=10, color='blue', alpha=0.5)
ax.scatter(np.log2(nuc_df['clip_enrichment']), np.log2(nuc_df['rden_change']), label='Nucleus', s=10, color='red', alpha=0.5)
ax.scatter(np.log2(ribo_df['clip_enrichment']), np.log2(ribo_df['rden_change']), label='Ribosome', s=10, color='green', alpha=0.5)

# 축 라벨 및 범례 설정
ax.set_xlabel('log2(clip_enrichment)')
ax.set_ylabel('log2(rden_change)')
ax.legend()

# Pearson 상관계수 표시
ax.text(0.95, 0.05, f'r = {r:.2f}', transform=ax.transAxes, ha='right', va='bottom')

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pysam

def plot_start_codon_density(bam_file_path, start_codon_bed_path, window_size=100):

  start_codon_positions = {}
  with open(start_codon_bed_path, 'r') as f:
    for line in f:
      fields = line.strip().split('\t')

      if len(fields) > 7:
          chrom, source, feature, start, end, score, strand, *_ = fields
      else:
          chrom, source, feature, start, end, score, strand = fields

      gene_id = fields[-1]

      start_codon_positions[gene_id] = (chrom, int(start), strand)


  density = {}
  for gene_id in start_codon_positions:
    density[gene_id] = np.zeros(2 * window_size + 1)


  with pysam.AlignmentFile(bam_file_path, "rb") as bamfile:
    for read in bamfile:
      if read.is_unmapped or read.is_secondary or read.is_duplicate:
        continue

      gene_id = read.reference_name
      if gene_id in start_codon_positions:
        chrom, start, strand = start_codon_positions[gene_id]


        if strand == '+':
          read_pos = read.reference_start - (start -1)+ window_size
        else:
          read_pos = (start -1)- read.reference_end + window_size


        if 0 <= read_pos <= 2 * window_size:
          density[gene_id][read_pos] += 1


  for gene_id, gene_density in density.items():
      if sum(gene_density) > 0:  # density가 0이 아닌 유전자
          #normalized_density = gene_density / sum(gene_density)
          #plt.plot(np.arange(-window_size, window_size + 1), normalized_density, alpha=0.5)

          plt.plot(np.arange(-window_size, window_size + 1), gene_density, alpha=0.5)


  plt.title('Ribosome Footprint Density near Start Codons')
  plt.xlabel('Distance from Start Codon (bp)')
  plt.ylabel('Density')
  plt.show()

bam_file_path = '../binfo1-datapack1/RPF-siLuc.bam'
start_codon_bed_path = 'gencode-start.gtf'
plot_start_codon_density(bam_file_path, start_codon_bed_path)
```
