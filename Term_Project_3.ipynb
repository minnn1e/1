```python
import pandas as pd

pileup = pd.read_csv('CLIP-let7g-gene.pileup', sep='\t', names=['chrom','pos','_ref','count','basereads','quals'])
pileup.tail()
```

```python
import re
toremove = re.compile('[<>$*#^]')
pileup['matches'] = pileup['basereads'].apply(lambda x: toremove.sub('',x))
```

```python
pileup[['chrom','pos','matches']]
```

```python
pileup[pileup['pos'] == 106056094].iloc[0]['matches']
```

```python
base_counts = pileup['matches'].apply(lambda x: len(x))
print("각 position별 base 수:")
print(base_counts.head())
```

```python
import math

def calculate_entropy(matches):
    base_freq = {}
    total = len(matches)
    for base in matches:
        if base in base_freq:
            base_freq[base] += 1
        else:
            base_freq[base] = 1

    entropy = 0
    for freq in base_freq.values():
        p = freq / total
        entropy += p * math.log2(p)
    return -entropy

# Shannon entropy 계산
pileup['entropy'] = pileup['matches'].apply(calculate_entropy)
print("\n각 position별 Shannon entropy:")
print(pileup[['pos', 'entropy']].head())
```

```python
# bedgraph
with open('entropy.bedgraph', 'w') as f:
    for idx, row in pileup.iterrows():
        f.write(f"{row['chrom']}\t{row['pos']-1}\t{row['pos']}\t{row['entropy']}\n")

print("\n생성된 bedgraph 파일의 내용:")
with open('entropy.bedgraph', 'r') as f:
    print(f.read())
```
